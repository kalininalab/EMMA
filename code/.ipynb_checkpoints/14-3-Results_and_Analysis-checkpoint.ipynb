```python
import numpy as np
from sklearn.metrics import confusion_matrix

# Function to calculate MCC and Accuracy
def calculate_metrics(y_true, y_pred):
    y_pred_binary = [1 if pred >= 0.5 else 0 for pred in y_pred]
    cm = confusion_matrix(y_true, y_pred_binary)
    TN, FP, FN, TP = cm.ravel()
    
    # Calculate MCC
    numerator = (TP * TN) - (FP * FN)
    denominator = np.sqrt((TP + FP) * (TP + FN) * (TN + FP) * (TN + FN))
    mcc = numerator / denominator if denominator > 0 else 0  # Handle division by zero
    
    # Calculate Accuracy
    accuracy = (TP + TN) / (TP + FP + FN + TN)
    return mcc, accuracy

# Dataset names and their corresponding true and predicted labels
datasets = [
    ('C2', y_test_true_xgboost_ESM2_ChemBERTa_C2_2S, y_test_pred_xgboost_ESM2_ChemBERTa_C2_2S),
    ('C1e', y_test_true_xgboost_ESM2_ChemBERTa_C1e_2S, y_test_pred_xgboost_ESM2_ChemBERTa_C1e_2S),
    ('C1f', y_test_true_xgboost_ESM2_ChemBERTa_C1f_2S, y_test_pred_xgboost_ESM2_ChemBERTa_C1f_2S),
    ('I1e', y_test_true_xgboost_ESM2_ChemBERTa_I1e_2S, y_test_pred_xgboost_ESM2_ChemBERTa_I1e_2S),
    ('I1f', y_test_true_xgboost_ESM2_ChemBERTa_I1f_2S, y_test_pred_xgboost_ESM2_ChemBERTa_I1f_2S)
]

# Calculate MCC and Accuracy for each dataset
results = []
for name, y_true, y_pred in datasets:
    mcc, accuracy = calculate_metrics(y_true, y_pred)
    results.append((name, mcc, accuracy))

# Create and display the results table
print(f"{'Model':<20} {'MCC':<10} {'Accuracy':<10}")
print("=" * 40)
for name, mcc, accuracy in results:
    print(f"{name:<20} {mcc:.4f}    {accuracy:.4f}")
```

```python
# Function to calculate FPR, FNR, and additional metrics
def calculate_metrics(y_true, y_pred):
    # Convert predicted probabilities to binary labels
    y_pred_binary = [1 if pred >= 0.5 else 0 for pred in y_pred]
    
    # Calculate confusion matrix and metrics
    cm = confusion_matrix(y_true, y_pred_binary)
    TN, FP, FN, TP = cm.ravel()
    
    fpr = round(FP / (FP + TN),3)
    fnr = round(FN / (FN + TP),3)
    
    return fpr, fnr, cm

# Define datasets
datasets = [
    ('C2', y_test_true_xgboost_ESM2_ChemBERTa_C2_2S, y_test_pred_xgboost_ESM2_ChemBERTa_C2_2S),
    ('C1e', y_test_true_xgboost_ESM2_ChemBERTa_C1e_2S, y_test_pred_xgboost_ESM2_ChemBERTa_C1e_2S),
    ('C1f', y_test_true_xgboost_ESM2_ChemBERTa_C1f_2S, y_test_pred_xgboost_ESM2_ChemBERTa_C1f_2S),
    ('I1f', y_test_true_xgboost_ESM2_ChemBERTa_I1f_2S, y_test_pred_xgboost_ESM2_ChemBERTa_I1f_2S),
    ('I1e', y_test_true_xgboost_ESM2_ChemBERTa_I1e_2S, y_test_pred_xgboost_ESM2_ChemBERTa_I1e_2S)
]

# Set up the subplots
n_rows = 3
n_cols = 2
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 18))
#fig.suptitle('Confusion Matrices with Metrics for Different split methods', fontsize=16)

# Loop through datasets and plot confusion matrices with additional metrics
for i, (name, y_true, y_pred) in enumerate(datasets):
    fpr, fnr, cm,
    row, col = divmod(i, n_cols)
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(ax=axes[row, col], cmap='Blues', colorbar=False)
    
    # Add more metrics to the title
    axes[row, col].set_title(
        f"Split Method: {name}\nFPR: {fpr:.2f}, FNR: {fnr:.2f}",
        fontsize=16
    )
    
    # Adjust tick labels and axes formatting
    axes[row, col].tick_params(axis='both', which='major', labelsize=14)
    axes[row, col].set_xlabel(axes[row, col].get_xlabel(), fontsize=14)
    axes[row, col].set_ylabel(axes[row, col].get_ylabel(), fontsize=14)

# Remove any empty subplots if necessary
for j in range(len(datasets), n_rows * n_cols):
    fig.delaxes(axes.flat[j])

# Adjust layout for better visualization
plt.subplots_adjust(hspace=0.9, wspace=0.00001)
plt.tight_layout(rect=[0, 0,0.75, 0.9])
path= "/Users/vahidatabaigi/Desktop/Thesis/thesis-template/Figures/CM-SIP.png"
plt.savefig(path)
plt.show()
```
