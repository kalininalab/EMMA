```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, roc_auc_score
import numpy as np
from pathlib import Path

# Set style (using one of your available styles)
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    'font.family': 'Arial',
    'font.size': 10,
    'axes.labelsize': 10,
    'axes.titlesize': 12,
    'legend.fontsize': 9,
    'figure.dpi': 300,
    'savefig.dpi': 300,
})

def plot_roc_curves(inter_y_true, inter_y_pred, sub_y_true, sub_y_pred, split_name, ax=None):
    """Plot ROC curves for both heads with proper subclass handling"""
    if ax is None:
        fig, ax = plt.subplots(figsize=(8, 6))
    
    # 1. Interaction Head (all samples)
    fpr_i, tpr_i, _ = roc_curve(inter_y_true, inter_y_pred)
    roc_auc_i = auc(fpr_i, tpr_i)
    ax.plot(fpr_i, tpr_i, color='#1f77b4', lw=2, 
            label=f'Interaction (AUC = {roc_auc_i:.3f})')
    
    # 2. Subclass Head (only interacting pairs)
    interacting_mask = (inter_y_true == 1)
    if interacting_mask.sum() > 0:  # Only plot if there are interacting pairs
        try:
            # Ensure binary classification
            if len(np.unique(sub_y_true[interacting_mask])) == 2:
                fpr_s, tpr_s, _ = roc_curve(sub_y_true[interacting_mask], 
                                           sub_y_pred[interacting_mask])
                roc_auc_s = auc(fpr_s, tpr_s)
                ax.plot(fpr_s, tpr_s, color='#ff7f0e', lw=2,
                       label=f'Subclass (AUC = {roc_auc_s:.3f})')
            else:
                print(f"Warning: Subclass head for {split_name} doesn't have exactly 2 classes")
        except ValueError as e:
            print(f"Could not plot subclass ROC for {split_name}: {str(e)}")
    
    # Formatting
    ax.plot([0, 1], [0, 1], 'k--', lw=1, alpha=0.5)
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate', fontweight='bold')
    ax.set_ylabel('True Positive Rate', fontweight='bold')
    ax.set_title(f'{split_name} split', fontweight='bold')
    ax.legend(loc="lower right")
    ax.grid(True, linestyle=':', alpha=0.3)
    
    return ax

# Create a figure with subplots for each split
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot each split
plot_roc_curves(inter_y_test_true_C1f, inter_y_test_pred_C1f,
                sub_y_test_true_C1f, sub_y_test_pred_C1f,
                'S1$_{p}$', ax=axes[0, 0])

plot_roc_curves(inter_y_test_true_C1e, inter_y_test_pred_C1e,
                sub_y_test_true_C1e, sub_y_test_pred_C1e,
                'S1$_{l}$', ax=axes[0, 1])

plot_roc_curves(inter_y_test_true_I2, inter_y_test_pred_I2,
                sub_y_test_true_I2, sub_y_test_pred_I2,
                'I2', ax=axes[1, 0])

plot_roc_curves(inter_y_test_true_C2, inter_y_test_pred_C2,
                sub_y_test_true_C2, sub_y_test_pred_C2,
                'S2', ax=axes[1, 1])

plt.tight_layout(pad=3.0)

# Save figure
output_path = Path(CURRENT_DIR) / ".." / "figures" / "auc_roc_by_split.png"
output_path.parent.mkdir(exist_ok=True)
plt.savefig(output_path, bbox_inches='tight', dpi=300)
# plt.close()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

def calculate_metrics(inter_y_true, inter_y_pred, sub_y_true, sub_y_pred):
    """Calculate metrics for both heads"""
    metrics = {}
    
    # Interaction head (binary classification)
    inter_pred_class = (inter_y_pred >= 0.5).astype(int)
    metrics['inter_accuracy'] = accuracy_score(inter_y_true, inter_pred_class)
    metrics['inter_f1'] = f1_score(inter_y_true, inter_pred_class)
    
    # Subclass head (only on interacting pairs)
    interacting_mask = (inter_y_true == 1)
    if interacting_mask.sum() > 0:
        sub_pred_class = (sub_y_pred[interacting_mask] >= 0.5).astype(int)
        if len(np.unique(sub_y_true[interacting_mask])) == 2:  # Ensure binary
            metrics['sub_accuracy'] = accuracy_score(
                sub_y_true[interacting_mask], sub_pred_class)
            metrics['sub_f1'] = f1_score(
                sub_y_true[interacting_mask], sub_pred_class)
    
    return metrics

# Calculate for all splits
splits = ['C1f', 'C1e', 'I2', 'C2']
results = {}

for split in splits:
    results[split] = calculate_metrics(
        globals()[f'inter_y_test_true_{split}'],
        globals()[f'inter_y_test_pred_{split}'],
        globals()[f'sub_y_test_true_{split}'],
        globals()[f'sub_y_test_pred_{split}']
    )
import pandas as pd


results_df = pd.DataFrame(results).T.round(3)
results_df.columns = ['Interaction Accuracy', 'Interaction F1', 
                     'Subclass Accuracy', 'Subclass F1']

# Display the table
print(results_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    'font.family': 'Arial',
    'font.size': 10,
    'axes.titlesize': 12,
    'figure.dpi': 300
})

# Prepare data
x = np.arange(len(splits))  # Split locations
width = 0.35  # Bar width

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Accuracy Plot
for i, split in enumerate(splits):
    ax1.bar(x[i] - width/2, results[split].get('inter_accuracy', np.nan), 
            width, label='Interaction' if i==0 else "", color='#1f77b4')
    ax1.bar(x[i] + width/2, results[split].get('sub_accuracy', np.nan), 
            width, label='Subclass' if i==0 else "", color='#ff7f0e')

ax1.set_xticks(x)
ax1.set_xticklabels(splits)
ax1.set_ylim(0, 1.05)
ax1.set_ylabel('Accuracy', fontweight='bold')
ax1.set_title('Accuracy by Split', fontweight='bold')
ax1.legend()

# F1 Score Plot
for i, split in enumerate(splits):
    ax2.bar(x[i] - width/2, results[split].get('inter_f1', np.nan), 
            width, label='Interaction' if i==0 else "", color='#1f77b4')
    ax2.bar(x[i] + width/2, results[split].get('sub_f1', np.nan), 
            width, label='Subclass' if i==0 else "", color='#ff7f0e')

ax2.set_xticks(x)
ax2.set_xticklabels(splits)
ax2.set_ylim(0, 1.05)
ax2.set_ylabel('F1 Score', fontweight='bold')
ax2.set_title('F1 Score by split', fontweight='bold')

plt.tight_layout()
plt.savefig('accuracy_f1_comparison.png', bbox_inches='tight', dpi=300)
# plt.close()
plt.show()
```
